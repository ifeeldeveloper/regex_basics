```python
apple_regex = 'apples'

matched_result = re.search(apple_regex, my_doc)
print(matched_result.group())
print(matched_result.start())
print(matched_result.end())
print(matched_result.span())
```

```python
date_regex = r'([12]\d{3})([-\/\.]?)(\d{2})\2(\d{2})'

matched_result = re.search(date_regex, my_doc)
print(matched_result.group())
print(matched_result.start())
print(matched_result.end())
print(matched_result.span())
```

```python
date_regex = r'([12]\d{3})([-\/\.]?)(\d{2})\2(\d{2})'

matched_result = re.search(date_regex, my_doc)
print(matched_result.group(1))
print(matched_result.start())
print(matched_result.end())
print(matched_result.span())
```

```python
date_regex = r'([12]\d{3})([-\/\.]?)(\d{2})\2(\d{2})'

matched_result = re.search(date_regex, my_doc)
print(matched_result.group(1,3,4))
print(matched_result.start())
print(matched_result.end())
print(matched_result.span())
```

```python
date_regex = r'(?P<Year>[12]\d{3})([-\/\.]?)(?P<Month>\d{2})\2(?P<Day>\d{2})'

matched_result = re.search(date_regex, my_doc)
print(matched_result.group("Year", "Month", "Day"))
print(matched_result.start())
print(matched_result.end())
print(matched_result.span())
```

```python
price_data_1 = '1,200'
price_data_2 = '2-100'

reg = r'[^\d]'

price_data_1_cleaned = re.sub(reg, '', price_data_1)
price_data_2_cleaned = re.sub(reg, '', price_data_2)

print(price_data_1_cleaned)
print(price_data_2_cleaned)
```

```python
price_data_1 = '1,200'
price_data_2 = '2-100'

reg = r'[-,]'

price_data_1_cleaned = re.sub(reg, '', price_data_1)
price_data_2_cleaned = re.sub(reg, '', price_data_2)

print(price_data_1_cleaned)
print(price_data_2_cleaned)
```
